Zero-shot Prompting

```python
from google import genai
from google.colab import userdata

# 1. Securely retrieve your API key and initialize the client
api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key)

# We are using the standard 3.5-flash model as it is fast and excellent at instruction following
MODEL_ID = "gemini-3.5-flash"

print("Running Zero-Shot Examples...\n" + "="*40 + "\n")

# ==========================================
# Example A: The Formatting Challenge
# ==========================================
# In a zero-shot prompt, you ask for a specific format (like JSON)
# without showing it what the output should look like beforehand.

prompt_a = """
Extract the name, occupation, and city from the following sentence and output it as JSON:

"My name is Sarah, I work as a mechanical engineer, and I just moved to Seattle."
"""

print("### Example A: The Formatting Challenge ###")
response_a = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt_a
)
print(response_a.text)
print("\n" + "="*40 + "\n")


# ==========================================
# Example B: Creative Constraint
# ==========================================
# Here, you are testing the model's ability to strictly follow
# structural rules (word counts) on the very first try.

prompt_b = """
Write a two-sentence horror story about a smartphone.
The first sentence must be exactly 5 words.
The second sentence must be exactly 3 words.
"""

print("### Example B: Creative Constraint ###")
response_b = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt_b
)
print(response_b.text)
```

One-Shot Prompting

```python
from google import genai
from google.colab import userdata

# 1. Securely retrieve your API key and initialize the client
api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key)

# Using the standard 3.5-flash model
MODEL_ID = "gemini-3.5-flash"

print("Running One-Shot Examples...\n" + "="*40 + "\n")

# ==========================================
# Example A: Tone Translation
# ==========================================
# By providing one clear example of translating jargon to plain English,
# the model instantly understands the "vibe" and level of simplicity you want.

prompt_a = """
Convert the corporate jargon into plain English.

Corporate: "Let's synergize our bandwidth to touch base on the deliverables."
Plain English: "Let's work together and meet to discuss the project."

Corporate: "We need to boil the ocean to shift a paradigm in this vertical."
Plain English:
"""

print("### Example A: Tone Translation ###")
response_a = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt_a
)
# The model will complete the prompt starting from "Plain English:"
print(response_a.text.strip())
print("\n" + "="*40 + "\n")


# ==========================================
# Example B: Strict Data Extraction
# ==========================================
# Instead of writing a long paragraph explaining what a pipe-separated
# format is, showing it one example perfectly locks in the structure.

prompt_b = """
Extract the flight details into a pipe-separated format.

Input: "I'm flying on Delta flight 402 from JFK to LAX on Tuesday."
Output: Delta | 402 | JFK | LAX | Tuesday

Input: "Book me on United 88 departing from ORD and arriving at SFO tomorrow."
Output:
"""

print("### Example B: Strict Data Extraction ###")
response_b = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt_b
)
# The model will follow the exact pipe-separated pattern for the new input
print(response_b.text.strip())
```

Few-shot prompting

```python
from google import genai
from google.colab import userdata

# 1. Securely retrieve your API key and initialize the client
api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key)

# We continue using the fast and capable 3.5-flash model
MODEL_ID = "gemini-3.5-flash"

print("Running Few-Shot Examples...\n" + "="*40 + "\n")

# ==========================================
# Example A: Custom Routing Logic
# ==========================================
# By providing multiple examples of different ticket types,
# the model learns the exact boundaries of your custom categories
# and the precise bracketed format you expect.

prompt_a = """
Classify the customer support ticket into one of three categories: [BILLING], [TECH_ISSUE], or [SALES].

Ticket: "My screen is cracked and the touch sensor won't work."
Category: [TECH_ISSUE]

Ticket: "Do you offer enterprise discounts for teams of 50 or more?"
Category: [SALES]

Ticket: "I was double-charged on my credit card this month."
Category: [BILLING]

Ticket: "How do I upgrade my account from basic to premium?"
Category:
"""

print("### Example A: Custom Routing Logic ###")
response_a = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt_a
)
# The model will complete the final category based on the pattern
print(response_a.text.strip())
print("\n" + "="*40 + "\n")


# ==========================================
# Example B: Grocery Categorization
# ==========================================
# Multiple examples help the model map real-world items (like 'Sourdough Loaf')
# to your specific, predefined system labels ([BAKERY]).

prompt_b = """
Categorize the grocery items into the correct department: [PRODUCE], [DAIRY], [BAKERY], or [MEAT].

Item: Granny Smith Apples
Department: [PRODUCE]

Item: Whole Milk
Department: [DAIRY]

Item: Sourdough Loaf
Department: [BAKERY]

Item: Ground Beef
Department: [MEAT]

Item: Organic Carrots
Department:
"""

print("### Example B: Grocery Categorization ###")
response_b = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt_b
)
# The model will correctly classify the final item
print(response_b.text.strip())
```

Chain-of-thought prompting

```python
from google import genai
from google.colab import userdata

# 1. Securely retrieve your API key and initialize the client
api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key)

# Using the standard 3.5-flash model, which handles reasoning tasks exceptionally well
MODEL_ID = "gemini-3.5-flash"

print("Running Chain-of-Thought Examples...\n" + "="*40 + "\n")

# ==========================================
# Example A: The Logic Puzzle
# ==========================================
# Without CoT, AI models often jump to the intuitive (but wrong) answer of "100 minutes".
# Forcing it to explain the steps ensures it realizes the rate is 1 widget per machine per 5 minutes.

prompt_a = """
Solve the following logic puzzle. Before giving the final answer, break down your reasoning step-by-step.

Puzzle: If it takes 5 machines 5 minutes to make 5 widgets, how long would it take 100 machines to make 100 widgets?
"""

print("### Example A: The Logic Puzzle ###")
response_a = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt_a
)
print(response_a.text)
print("\n" + "="*40 + "\n")


# ==========================================
# Example B: Schedule Resolution
# ==========================================
# Scheduling requires comparing overlapping windows. By asking it to think
# step-by-step, the model will physically write out the overlaps,
# preventing it from guessing a time that excludes one person.

prompt_b = """
Let's find a meeting time. Think through the constraints step-by-step before giving the final answer.

Constraints:
* Alice is available from 9:00 AM to 11:30 AM.
* Bob is available from 10:00 AM to 12:00 PM.
* Charlie is available from 10:30 AM to 2:00 PM.
* The meeting needs to last exactly 45 minutes.

What is the earliest possible time they can all meet?
"""

print("### Example B: Schedule Resolution ###")
response_b = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt_b
)
print(response_b.text)
```

Demo: The Temperature Dial

What the Class Will Observe

At Temperature 0.0: The outputs will be nearly (or completely) identical every single time. It will likely output something safe and highly predictable like, "Your daily grind, perfected." or "Fueling your day, one cup at a time." Explain to the class that at 0.0, the model is acting like a strict calculator, picking the most mathematically probable next word.

At Temperature 1.0: The outputs will diverge wildly. You will get five distinct taglines with different tones, vocabulary, and angles. Explain that higher temperatures introduce randomness, allowing the model to pick less probable—but more interesting—words.

Pro-Tip for the Classroom: If you want to show truly wild variations (sometimes bordering on nonsensical or bizarre), change the second temperature setting from 1.0 to 2.0 (the maximum allowed). It is a great way to show what happens when the AI is given too much "creative freedom."

```python
import time
from google import genai
from google.genai import types
from google.colab import userdata

# 1. Securely retrieve your API key and initialize the client
api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key)

MODEL_ID = "gemini-3.5-flash"
PROMPT = "Write a tagline for a coffee shop."

print("☕ THE TEMPERATURE DIAL DEMO ☕\n" + "="*40)
print(f"Prompt: '{PROMPT}'\n")

# ==========================================
# Test 1: Temperature 0.0 (The Robot)
# ==========================================
print("🌡️ RUNNING AT TEMPERATURE 0.0 (Highly Deterministic / Focused)")
print("Observe: The model should pick the most statistically likely answer and stick to it.\n")

for i in range(1, 6):
    response = client.models.generate_content(
        model=MODEL_ID,
        contents=PROMPT,
        config=types.GenerateContentConfig(
            temperature=0.0,
        )
    )
    print(f"Attempt {i}: {response.text.strip()}")
    time.sleep(1) # Small pause to avoid API rate limits

print("\n" + "-"*40 + "\n")

# ==========================================
# Test 2: Temperature 1.0 (The Creative)
# ==========================================
print("🔥 RUNNING AT TEMPERATURE 1.0 (Highly Creative / Varied)")
print("Observe: The model is allowed to take risks, resulting in different ideas every time.\n")

for i in range(1, 6):
    response = client.models.generate_content(
        model=MODEL_ID,
        contents=PROMPT,
        config=types.GenerateContentConfig(
            temperature=1.0,
        )
    )
    print(f"Attempt {i}: {response.text.strip()}")
    time.sleep(1)

print("\n" + "="*40)
print("Demo Complete! Have the class discuss the differences.")
```

The Raw JSON Demo

```python
import os
import json
from google import genai

# 1. Ensure you have your API key set in your environment variables
# export GEMINI_API_KEY="your_api_key_here"

def run_json_demo():
    print("Initializing Gemini Client...")
    # The client automatically picks up the GEMINI_API_KEY environment variable
    api_key = userdata.get('GEMINI_API_KEY')
    client = genai.Client(api_key=api_key)

    prompt = "Why is the sky blue? Answer in one short sentence."
    print(f"Sending prompt: '{prompt}'\n")

    # 2. Make the API call
    response = client.models.generate_content(
        model='gemini-3.5-flash',
        contents=prompt,
    )

    # 3. Print the standard text output (what the user usually sees)
    print("=== WHAT THE USER SEES ===")
    print(response.text)
    print("\n")

    # 4. Print the raw JSON structure (what the machine actually receives)
    print("=== WHAT THE MACHINE SEES (RAW JSON) ===")

    # The response object is a Pydantic model, so we can easily dump it to JSON
    raw_json = response.model_dump_json(indent=2)
    print(raw_json)

if __name__ == "__main__":
    run_json_demo()
```
